```python
X = df.drop(['charges'],axis=1)
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
encoding = ColumnTransformer([
    ('Ohe',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[1,3,4,5])
],remainder='passthrough')
```

```python
scaling= ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,12))
],remainder='passthrough')
```

```python
model = LinearRegression()
```

```python
pipe= Pipeline([
    ('encoding',encoding),
    ('scaling',scaling),
])
```

```python
X_train
```

```python
pipe.fit(X_train)
```

```python
encoded_X_train= pipe.transform(X_train)
encoded_X_test=pipe.transform(X_test)
```

```python
encoded_X_train
```

```python
model = LinearRegression()
model.fit(encoded_X_train,y_train)
```

```python
y_pred1 = model.predict(encoded_X_test)
score1 = metrics.r2_score(y_test,y_pred1)
mae1 = metrics.mean_absolute_error(y_test,y_pred1)

print("score ",score1)
print("mean absolute error ",mae1)
```

```python
model = SVR()
model.fit(encoded_X_train,y_train)
```

```python
y_pred2 = model.predict(encoded_X_test)
score2 = metrics.r2_score(y_test,y_pred2)
mae2 = metrics.mean_absolute_error(y_test,y_pred2)

print("score ",score2)
print("mean absolute error ",mae2)
```

```python
model = RandomForestRegressor()
model.fit(encoded_X_train,y_train)
```

```python
y_pred3 = model.predict(encoded_X_test)
score3 = metrics.r2_score(y_test,y_pred3)
mae3 = metrics.mean_absolute_error(y_test,y_pred3)

print("score ",score3)
print("mean absolute error ",mae3)
```

```python
model = DecisionTreeRegressor()
model.fit(encoded_X_train,y_train)
```

```python
y_pred4 = model.predict(encoded_X_test)
score4 = metrics.r2_score(y_test,y_pred4)
mae4 = metrics.mean_absolute_error(y_test,y_pred4)

print("score ",score4)
print("mean absolute error ",mae4)
```

```python
model = GradientBoostingRegressor()
model.fit(encoded_X_train,y_train)
```

```python
y_pred5= model.predict(encoded_X_test)
score5 = metrics.r2_score(y_test,y_pred5)
mae5 = metrics.mean_absolute_error(y_test,y_pred5)
print("score ",score5)
print("mean absolute error ",mae5)
```

```python
final= Pipeline([
    ('encoding',encoding),
    ('scaling',scaling),
    ('model',model)
])
```

```python
final.fit(X_train,y_train)
```

```python
import pickle
pickle.dump(final,open('hicp.pkl', 'wb'))
```
